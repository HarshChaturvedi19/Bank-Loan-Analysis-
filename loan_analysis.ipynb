```python
# change the dtype of date format 
df['issue_date'].dtype
```

```python
df['issue_date'] = pd.to_datetime(
    df['issue_date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
df['issue_date'].dtype
```

```python
# convert the last credit pull date 
df['last_credit_pull_date'] = pd.to_datetime(
    df['last_credit_pull_date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
# convert the last_payment_date
df['last_payment_date'] = pd.to_datetime(
    df['last_payment_date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
# convert the next_payment_date
df['next_payment_date'] = pd.to_datetime(
    df['next_payment_date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
# check if there are any invalid or missing dates 
df[['issue_date','last_credit_pull_date','last_payment_date','next_payment_date']].isnull().sum()
```
